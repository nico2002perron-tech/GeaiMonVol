### 1. Introduction

La boucle `while` continue de boucler tant que sa condition est **True** (d'où son nom), mais que se passe-t-il si vous ne souhaitez exécuter un bloc de code qu'un **certain nombre de fois ?** 

Vous pouvez le faire avec une instruction de `boucle for` et la fonction `range()`.

Dans le code, une instruction `for` ressemble à quelque chose comme `for i in range(5)`: et inclut les éléments suivants :
- Le mot clé `for`
- Un nom de variable, par exemple *i*
- Le mot clé `in`
- Un appel à la méthode `range()` avec jusqu'à trois entiers qui lui sont passés
- Deux points `:`
- À partir de la ligne suivante, un bloc de code indenté (appelé la clause for)

En général, `une boucle for` est utilisée pour itérer sur une séquence (c'est-à-dire une liste, un tuple, un dictionnaire, un ensemble ou une chaîne).

Mais avant de continuer sur la boucle `for`, parlons d'abord de la fonction `range()`.

### 2. La fonction `range()`:

La fonction `range()` renvoie une séquence de nombres, commençant par 0 par défaut, et incrémentée de 1 (par défaut), et s'arrête avant un nombre spécifié par exemple range(5) renvoie la séquence : 0, 1, 2, 3, 4 <u> mais  pas 5</u>.

La syntaxe générale de la fonction range est la suivante :

**range(start, stop, step)** par exemple : **range(3, 30, 4)**


|Paramètre|Description|Exemple|
|:---|:---|:---:|
|start |Facultatif. Un nombre entier spécifiant à quelle position commencer. La valeur par défaut est 0 |3| 
|stop|  **Obligatoire**. Un nombre entier spécifiant à quelle position s'arrêter (**non inclut dans la séquence**). |30| 
|step |  Facultatif. Un nombre entier spécifiant l'incrémentation. La valeur par défaut est 1 |4| 



**2.1. Créez une séquence de nombres de 0 à 5 et imprimez chaque élément de la séquence :**

```python
#Quand un seul parmètre dans la fonction, on commence par défaut par 0 
#jusqu'à 5, le dernier chiffre n'est jamais inclut, dans ce cas 6.

x = range(6)

for n in x:
    print(n)
```

**2.2 Créez une séquence de nombres de 3 à 5 et imprimez chaque élément de la séquence :**

```python
#quand deux parmètres dans la fonction, on commence par le premier paramètre 3 
#jusqu'au deuxième paramètre 6. Remarquez encore que 6 n'est pas inclut

x = range(5, 10)

for n in x:
    print(n)
```

**2.3. Créez une séquence de nombres de 3 à 29 avec un pas de 2 et imprimez chaque élément de la séquence :**

```python
#quand trois parmètres dans la fonction, le troisième représente toujours le pas (step) 

for n in range(4, 26, 3):
    print(n)
```

**2.4. Utiliser `range()` pour créer une liste de nombres:**

Si vous souhaitez créer une liste de nombres, vous pouvez convertir les résultats de range() directement en une liste à l'aide de la fonction list(). 

Dans les exemples de la section précédente, nous avons simplement imprimé une série de nombres. Nous pouvons utiliser list() pour convertir ce même ensemble de nombres en une liste :

```python
print(range(6))
```

```python
liste1 = list(range(6))
liste2 = list(range(3, 6))
liste3 = list(range(3, 30, 2))

print("liste1 :", liste1)
print("liste2 :", liste2)
print("liste3 :", liste3)
```

### 3.   Exemples de la boucle for:

#### 3.1. Créer une liste de carrés:

```python
liste_carrés = []                 # déclarer une variable avec une liste vide
    
for i in range(15):            # on choisit par ordre un élément de la séquence range(1,15)
    carré = i**2                  # calcul du carré de chaque éléement dans la séquence range(1, 15) 
    liste_carrés.append(carré)    # on ajoute la carré calculé à la laiste liste_carrés
    
print(liste_carrés)

print(len(liste_carrés))
    
```

Pour écrire ce dernier code de manière plus concise, omettez la variable temporaire `carré` et ajoutez chaque nouvelle valeur directement à la liste :

```python
liste_carrés = []                 # déclarer une variable avec une liste vide

for i in range(1, 15):            # on choisit par ordre un élément de la séquence range(1,15)
    liste_carrés.append(i**2)     # on ajoute directement la carré calculé à la laiste liste_carrés
    
print(liste_carrés)
    
```

#### 3.2. List Comprehensions  ou liste en compréhensions:

L'approche décrite précédemment pour générer les cases de la liste consistait à utiliser trois ou quatre lignes de code. 

Une compréhension de liste permet de générer cette même liste en une seule ligne de code. 

Une compréhension de liste combine `la boucle for` et la création de nouveaux éléments sur une seule ligne, et ajoute automatiquement chaque nouvel élément. 

Les compréhensions de liste ne sont pas toujours présentées aux débutants, mais je les ai incluses ici parce que vous les verrez très probablement dès que vous commencerez à regarder le code des autres.

L'exemple suivant construit la même liste de nombres carrés que vous avez vue précédemment, mais utilise une compréhension de liste :

```python
liste_carrés = [i**2 for i in range(1, 15)]

print(liste_carrés)
```

#### 3.3. Imprimer les éléments d'une liste avec la boucle for

```python
big_techs = ['Google', 'Apple', 'Amazon', 'Facebook', 'Microsoft']

for i in range(len(big_techs)):
    print(big_techs[i])
```

### 4. L'organigramme de `la boucle for`:

- Le prgramme dans la cellule suivante exécute `une boucle for` cinq fois. 

- La première fois qu'il est exécuté, la variable i est prend la valeur 0.

- Ensuite, la fonction print() dans la clause affichera : **Benoît cinq fois (0)**. 
- Une fois que Python a terminé une itération dans tout le code à l'intérieur de la clause de la boucle for, l'exécution revient au début de la boucle et l'instruction for incrémente i de un.
- C'est pourquoi range(5) donne cinq itérations dans la clause, i étant défini de 0, puis 1, puis 2, puis 3, puis 4. La variable i ira jusqu'à, mais n'inclura pas, l'entier passé à range() c'est-à-dire 5. 


```python
print("Mon nom est :")

for i in range(5):
    print(f"Benoît cinq fois ({i+1})")
```

### 5. Les instructions  `break` et `continue`

En Python, les instructions `break` et `continue` peuvent modifier le déroulement d'une boucle normale (boucle while ou boucle for).

Les boucles parcourent un bloc de code jusqu'à ce que la condition de test soit **False**, mais parfois nous souhaitons terminer l'itération en cours ou même la boucle entière sans vérifier l'expression de la condition.

Les instructions `break` et `continue` sont utilisées dans ces cas.

#### 5.1 L'instruction `break`:

L'instruction `break` **termine** la boucle qui la contient. Le contrôle du programme passe à l'instruction immédiatement après le corps de la boucle qui vient d'être arrêtée.

Si l'instruction `break` est à l'intérieur d'une boucle imbriquée (boucle à l'intérieur d'une autre boucle), l'instruction `break` terminera la boucle la plus interne.

Syntaxe de break : `break`

Voici des exemples de l'instruction `break` :


```python
for lettre in "caractères":
    print(lettre)

```

Dans ce dernier programme, nous parcourons la séquence "caractères". Nous vérifions si la lettre est "t", sur laquelle nous sortons de la boucle. Par conséquent, nous voyons dans notre sortie que toutes les lettres jusqu'à "t" sont imprimées. Après cela, la boucle se termine.

Voici un autre exemple avec une liste cette fois-ci :

```python
big_techs = ['Google', 'Apple', 'Amazon', 'Facebook', 'Microsoft']

for firme in big_techs:
    if firme == "Facebook":
        break
    print(firme)

print("\nFin de la boucle sans imprimer le reste des entreprises.")
```

#### 5.2 L'instruction `continue`:

L'instruction continue est utilisée pour **ignorer** le reste du code à l'intérieur d'une boucle **pour l'itération en cours uniquement**. La boucle ne se termine pas mais continue avec l'itération suivante.

Syntaxe de l'instruction continuer : `continue`

Voici des exemples de l'instruction `continue` :

```python
for lettre in "caractères":
    if lettre == "t":
        continue
    print(lettre)

print("\nRemaqruez que la lettre 't' est ignorée dans la boucle.")
```

Ce dernier programme est identique à l'exemple ci-dessus sur l'instruction `break`, sauf que l'instruction `break` a été remplacée par `continue`.

Quand la variable lettre est égale à "t", nous continuons avec la boucle sans exécuter l'instruction pour la lettre "t". Par conséquent, nous voyons dans notre sortie que toutes les lettres sont imprimées sauf la lettre "t".

Même chose pour le programme suivant avec une liste. Remarquez que "facebook" n'est pas imprimée :

```python
big_techs = ['Google', 'Apple', 'Amazon', 'Facebook', 'Microsoft']

for firme in big_techs:
    if firme == "Facebook":
        continue
    print(firme)

print("\nRemarquez que 'Facebook' est ignorée dans la boucle.")
```
